```python
image = cv2.imread('540Q3nNHPHnHsZvnqw8mcH77vt8tg9eDyQG1pHUzLCQJ0uEZ6R4gso4sj5PjKU9f5Ytwmenu6mY4O0fdex_Y8JXi0njZl2pvB2ciFPRs7jiUtWlphHHiHbB4Yrh_EhHKDH-D3GVj_GqjNfu1I44dw9f2icmUEdCm2Q6RfeQFvIs.jpg')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(7,7))
plt.imshow(image)
plt.title("Original CCTV Image")
plt.axis("off")
plt.show()
```

```python
print("Image Shape :", image.shape)
print("Height :", image.shape[0])
print("Width :", image.shape[1])
print("Channels :", image.shape[2])
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
equalized = cv2.equalizeHist(gray)

plt.figure(figsize=(6,6))
plt.imshow(equalized, cmap='gray')
plt.title("Histogram Equalization")
plt.axis("off")
plt.show()
```

```python
gaussian = cv2.GaussianBlur(equalized,(5,5),0)

plt.imshow(gaussian,cmap='gray')
plt.title("Gaussian Filter")
plt.axis("off")
plt.show()
```

```python
median = cv2.medianBlur(gaussian,5)

plt.imshow(median,cmap='gray')
plt.title("Median Filter")
plt.axis("off")
plt.show()
```

```python
bright = cv2.convertScaleAbs(median,
                             alpha=1.4,
                             beta=40)

plt.imshow(bright,cmap='gray')
plt.title("Brightness Enhanced Image")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(18,10))

titles = ["Original",
          "Gray",
          "Histogram Equalization",
          "Gaussian Filter",
          "Median Filter",
          "Brightness Enhanced"]

images = [image,
          gray,
          equalized,
          gaussian,
          median,
          bright]

for i in range(len(images)):
    plt.subplot(2,3,i+1)

    if len(images[i].shape)==2:
        plt.imshow(images[i], cmap='gray')
    else:
        plt.imshow(images[i])

    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
